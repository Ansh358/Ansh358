```python
# Scatter plot of Toss Winners in Team 1
plt.figure(figsize=(10, 6))
plt.scatter(df['team1'], df['toss_winner'], color='Orange', alpha=0.6, edgecolor='black')
plt.title('Scatter plot of Toss Winners in Team 1')
plt.xlabel('team1')
plt.ylabel('toss_winner')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Bar plot of Wining Team
plt.figure(figsize=(12, 6))
Wining_Team = df['team1'].value_counts().head(10)
sns.barplot(x=Wining_Team.index, y=Wining_Team.values, palette='viridis')
plt.title('Wining Team')
plt.xlabel('Wining Team')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
