```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('archive4.zip')
```

```python
plt.figure(figsize=(10,6))
df['Absences'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Chart for Absencies')
plt.xlabel('Absents')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Absences', y='Age', data=df, color='green')
plt.title('Scatterplot for age and absence  ')
plt.xlabel('Absences')
plt.ylabel('Age')
plt.show()

```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 6))

sns.histplot(df['Age'], bins=40, kde=True, color='skyblue', ax=axes[0])
axes[0].set_title('Age')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Count')

sns.histplot(df['StudyTimeWeekly'], bins=40, kde=True, color='green', ax=axes[1])
axes[1].set_title('StudyTimeWeekly')
axes[1].set_xlabel('StudyTimeWeekly')
axes[1].set_ylabel('Count')

sns.histplot(df['GPA'], bins=40, kde=True, color='orange', ax=axes[2])
axes[2].set_title('GPA')
axes[2].set_xlabel('GPA')
axes[2].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
